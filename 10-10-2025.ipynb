```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#create a sample coorelation matrix
data=np.random.rand(6,6)
corr=pd.DataFrame(data,columns=list("abcdef")).corr()
#plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidth=0.5)
plt.title("heatmap example-correlation matrix")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load example dataset
df=sns.load_dataset("iris")
#pairplot with color by species
sns.pairplot(df,hue="species",diag_kind="kde",palette="viridis")
plt.title("pairplot example-iris dataset",y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load example dataset
tips=sns.load_dataset("tips")
#barplot
plt.figure(figsize=(6,4))
sns.barplot(x="day",y="total_bill",data=tips,palette="muted")
plt.title("barplot example")
plt.show()
#boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x="day",y="total_bill",data=tips,palette="viridis")
plt.title("boxplotexample")
plt.show()
#violinplot
plt.figure(figsize=(6,4))
sns.violinplot(x="day",y="total_bill",data=tips,palette="viridis")
plt.title("violinplot example")
plt.show()
#countplot
plt.figure(figsize=(6,4))
sns.countplot(x="day",data=tips,palette="coolwarm")
plt.title("countplot example")
plt.show()
```

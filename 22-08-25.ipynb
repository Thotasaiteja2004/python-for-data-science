```python
import numpy as np
#1d array
arr1=np.array([1,2,3,4])
print("1d array:",arr1)
#2d array
arr2=np.array([[1,2,3],[4,5,6]])
print("2d array:\n",arr2)
#zero array
zero_arr=np.zeros((3,3))
print("zero array:\n",zero_arr)
#ones array
ones_arr=np.ones((2,4))
print("ones array:\n",ones_arr)
#range of numbers
range_arr=np.arange(0,10,2)#
#start,stop,Step # This line is causing the error
print("range array:",range_arr)
#random array
rand_arr=np.random.rand(3,3)
print("random array:\n",rand_arr)
```

```python
import numpy as np
arr=np.array([[1,2,3,4.5],[5,6,7,8]])

print("Shape:",arr.shape)
print("dimension:",arr.ndim)
print("data type:",arr.dtype)
print("size:",arr.size)
print("item size:",arr.itemsize)
#slicing
arr1 = np.array([10,20,30,40])
print("first element",arr1[0])

print("first element",arr1[-1])
print("slice[1:4]",arr1[1:4])
print("every second element",arr1[::2])

#indexing
arr2 = np.array([[1,2,35],[5,9,8]])
print("element at row 1, col 2",arr2[0,1])
print("second row",arr2[1,:])
print("second column",arr2[:,1])
```

```python
#mathematical functions
arr=np.array([0, np.pi/2, np.pi/3, np.pi])
print("sin values",np.sin(arr))
print("cos values",np.cos(arr))
print("cosine:",np.sqrt([1,4,9]))
print("exponential:",np.exp([1,2]))
print("logaritham:",np.log([1,np.e,np.e**2]))
```

```python
#aggregate fuctions
arr=np.array([[1,2,3],[4,5,6]])
print("sum:",np.sum(arr))
print("min:",np.min(arr))
print("max:",np.max(arr))
print("mean:",np.mean(arr))
print("standard deviation",np.std(arr))
print("sum along axis 0:",np.sum(arr,axis=0))  #column-wise sum
print("sum along axis 1:",np.sum(arr,axis=1))  #row-wise sum
```

```python
#broadcasting
a=np.array([1,2,3])
b=5
print("add scalar to array:",a+b)#[6 7 8]
c=np.array([[1],[2],[3]])
d=np.array([4,5,6])
print("broadcasting:\n",c+d)
```

```python
#numerical computing basic with numpy
#matrix operations
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
print("matrix multiplication:\n",np.dot(a,b))
print("matrix transpose:\n",np.transpose(a))
print("matrix inverse:\n",np.linalg.inv(a))
print("determinant:",np.linalg.det(a))
```

```python
#student marks analsis
import numpy as np
#marks of students in 3 subjects
marks=np.array([
    [85,90,88],
    [78,81,85],
    [92,89,95]
])
print("Average marks per student:", np.mean(marks, axis=1))
print("Average marks per subject:", np.mean(marks, axis=0))
print("Highest marks in each subject:", np.max(marks, axis=0))
print("Lowest marks in each subject:", np.min(marks, axis=0))
```
